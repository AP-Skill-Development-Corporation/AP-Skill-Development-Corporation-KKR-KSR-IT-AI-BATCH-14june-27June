```python
n=int(input()) # 3
f_count=0
for i in range(1,n+1): # 1   2   3
    if(n%i==0):    # 3%3==0(T)
        f_count=f_count+1 # 0+1=1 1+1=2
if(f_count==2): # 2==2(F)
    print("prime number")
else:
    print("Not a prime number")
```

```python
# Perfect number
# factors sum=given number
# 6= 1 2 3= 1+2+3=6

n=int(input()) #6
fsum=0    
for i in range(1,n): #1,5    2 3 4 5
    if(n%i==0): # 6%5==0(F)
        fsum=fsum+i #0+1=1 1+2=3  3+3=6 
if(fsum==n): # 6==6(T)
    print("perfect")
else:
    print("not perfect")
```
